# 9.6 示例： 树回归与标准回归的比较

## 创建一个回归树， 进行预测， 并计算预测值和实际值的相关系数

```python
trainMat = np.mat(regTrees.loadDataSet('bikeSpeedVsIq_train.txt'))
testMat = np.mat(regTrees.loadDataSet('bikeSpeedVsIq_test.txt'))
myTree = regTrees.createTree(trainMat, ops=(1,20))
yHat = regTrees.createForeCast(myTree, testMat[:,0])
print( "The correlation coefficient between yHat and yReal in the regression tree is " + 
      str(np.corrcoef(yHat, testMat[:,1], rowvar=0)[0,1]))
```

## 创建一个模型树， 进行预测， 并计算预测值和实际值的相关系数

```python
myTree = regTrees.createTree(trainMat, regTrees.modelLeaf, regTrees.modelErr, ops=(1,20))
yHat = regTrees.createForeCast(myTree, testMat[:,0], regTrees.modelTreeEval )
print( "The correlation coefficient between yHat and yReal in the model tree is " + 
      str(np.corrcoef(yHat, testMat[:,1], rowvar=0)[0,1]))
```

## 创建一个标准线性回归， 进行预测， 并计算预测值和实际值的相关系数

```python
ws, X, Y = regTrees.linearSolve(trainMat)
for i in range(np.shape(testMat)[0]):
    yHat[i] = testMat[i,0]*ws[1,0] + ws[0,0]
print( "The correlation coefficient between yHat and yReal in the linear model is " + 
      str(np.corrcoef(yHat, testMat[:,1], rowvar=0)[0,1]))
```
