# 8.1用线性回归找到最佳拟合曲线

```python
import regression
from numpy import *
```

```python
xArr, yArr = regression.loadDataSet('ex0.txt')
```

```python
xArr[0:3]
```

```python
ws = regression.standRegres(xArr, yArr)
print(ws)
```

```python
# Prediction
xMat = mat(xArr)
yMat = mat(yArr)
yHat = xMat*ws # predicted y
```

```python
# Plot
% matplotlib inline
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(xMat[:,1].flatten().A[0], yMat.T[:,0].flatten().A[0])

xCopy = xMat.copy()
# 将数据按升序排列
xCopy.sort(0)
yHat = xCopy*ws
ax.plot(xCopy[:,1], yHat)
plt.show()
```

```python
# 计算预测的 y 和真实的 y 的相关系数
yHat = xMat*ws
corrcoef(yHat.T, yMat)
```

# 8.3 示例：预测鲍鱼的年龄

```python
abX, abY = regression.loadDataSet('abalone.txt')
# Train
yHat01 = regression.lwlrTest(abX[0:99], abX[0:99], abY[0:99], 0.1)
yHat1 = regression.lwlrTest(abX[0:99], abX[0:99], abY[0:99], 1)
yHat10 = regression.lwlrTest(abX[0:99], abX[0:99], abY[0:99], 10)
print("Train: The residual sum of squares with k = " + str(0.1) + " is: " + str(regression.rssError(abY[0:99], yHat01.T)))
print("Train: The residual sum of squares with k = " + str(1) + " is: " + str(regression.rssError(abY[0:99], yHat1.T)))
print("Train: The residual sum of squares with k = " + str(10) + " is: " + str(regression.rssError(abY[0:99], yHat10.T)))

# Test
yHat01 = regression.lwlrTest(abX[100:199], abX[0:99], abY[0:99], 0.1)
yHat1 = regression.lwlrTest(abX[100:199], abX[0:99], abY[0:99], 1)
yHat10 = regression.lwlrTest(abX[100:199], abX[0:99], abY[0:99], 10)
print("Test: The residual sum of squares with k = " + str(0.1) + " is: " + str(regression.rssError(abY[100:199], yHat01.T)))
print("Test: The residual sum of squares with k = " + str(1) + " is: " + str(regression.rssError(abY[100:199], yHat1.T)))
print("Test: The residual sum of squares with k = " + str(10) + " is: " + str(regression.rssError(abY[100:199], yHat10.T)))
```
