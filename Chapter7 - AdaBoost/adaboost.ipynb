# 第七章： 利用 AdaBoost 元算法提高分类性能

## 7.3 基于单层决策树构建弱分类器

```python
import adaboost
import numpy as np
```

```python
datMat, classLabels = adaboost.loadSimpData()
```

```python
D = np.mat(np.ones((5,1))/5)
adaboost.buildStump(datMat, classLabels, D)
```

# 7.4 完整 AdaBoost 算法的实现

```python
classifierArray, aggClassEst = adaboost.adaBoostTrainDS(datMat, classLabels, 9)
```

# 7.5 测试算法: 基于AdaBoost的分类

```python
adaboost.adaClassify([0, 0], classifierArray)
```

```python
adaboost.adaClassify([0, 5], classifierArray)
```

```python
adaboost.adaClassify([5, 5], classifierArray)
```

# 7.6 示例: 在一个难数据集上应用AdaBoost

```python
datArr, labelArr = adaboost.loadDataSet('horseColicTraining2.txt')
classifierArray, aggClassEst = adaboost.adaBoostTrainDS(datArr, labelArr, 10)
```

```python
testArr, testLabelArr = adaboost.loadDataSet('horseColicTest2.txt')
prediction10 = adaboost.adaClassify(testArr, classifierArray)
# To get the number of misclassified examples
errArr = np.mat(np.ones((67,1)))
errArr[prediction10 != np.mat(testLabelArr).T].sum()
```
